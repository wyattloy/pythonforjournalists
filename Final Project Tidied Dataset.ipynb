## 2020 CO2 Emissions Data
### Summary
This dataset represents CO2 emissions data from the state of Ohio in 2020. It's part of a larger Excel workbook that outlines CO2 emissions data from all 50 states in 2020. I also have Excel workbooks for CO2 emissions data of every country in the world over the course of 2020.
#### [Link to Dataset(s)](https://www.icos-cp.eu/gcp-covid19)
#### [Link to Tidied Dataset](https://docs.google.com/spreadsheets/d/1C3BK5opFFYvyXKqYx-TGAdrC7Tes-Ogc89MBaK7DpVs/edit?usp=sharing)
### How I Cleaned the Data
I imported the data into Google Sheets, deleted all the unnecessary columns (for example, there were two seperate columns just saying "Ohio" and "OH" over and over; I deleted those) and changed the names of the columns so they read easier (for example, changing "AVI_CO2_MED" to "Aviation Median").
### Questions and Support for Story
I asked this data how CO2 emissions changed over the course of the pandemic in key sectors like aviation and surface transport (cars, busses, trains etc.). I found that it did fall significantly for a large chunk of the year, but then started to come back up.

These findings support my story in that it gives me a way to localize it. This topic seems hard to narrow down in a way that makes an Ohioan want to care--and it is hard--but this little bit of data can help me do that.
